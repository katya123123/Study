Есть 7 основ тестирования.

Чек-лист - это список проверок для тестирования продукта. 

Тест-кейс - набор предварительных условий, входных данных, действий, ожидаемых результатов, разработанных на основе условий тестирования. Это крайне точная и предельно понятная инструкция по поводу найденной ошибки. 

Результаты прохождения кейсов:
PASSED - тест пройден успешно, результат совпал с фактическими. Негативные тесты также могут быть пройденными.
FAILED - тест не пройден, ожидаемый результат НЕ совпал с фактическим.
QUESTION - кейс неясен, есть вопросы к заказчику.
BLOCKED - есть блокирующие проблемы, не позволяющие пройти кейс.

Модели разработки ПО:
1. Каскадная (ватерфолл). Каждый этап разработки идёт последовательно. Очень легко просчитать сроки. ПЛюсы: требования известны, четко прочитаны сроки. Минусы: невозможно внести изменения, любые именения вносить долго, тестирование начинается поздно. Ввоенной сфере или маленьких стартапах.
2. V-модель. На каждом этапе происходит валидация и тестирование. В сеферах науки, медицины, безопасности. Плюсы: требования зафиксированы, тщательное тестирование, каждая стадия имеет конкретный результат.
3. Инкриментная модель. ПЛюсы: дешевле менять требования, получение функционального продукта, упрощение тестирования. Минусы: проект полностью должен быть разработан в начале, необходимо очень четко понять интерфесы, тестирование на завершающей стадии.
4. Agile-методология. Суть в спринтах. Плюсы: гибкое управление, минимизируется число дефектов, легко реагирует на изменения. Минусы: постоянная обратная связь заказчика,необходимо адаптировать документацию, необходимость в частых встречах. 

Виды тестовой докумментации: 
1. Чек-лист. Список проверок для тестирования продукта. 
2. Тест-кейс - набор предварительных условий, входных действий, ожидаемых результатов и постусловий, разработанных на основе тестирования.
3. Тест-сьют-набор тестовых сценариев (тест-кейсов), которые будут использоваться в тестовом запуске.
4. Тест-план - докумнтация, описывающая цели тестирования, которые должны быть достигнуты, а также средства и графики их достижения, организованная дл координации деятельости тестирования + критерии выхода из тестирования. Работает только если поддерживается в актуальном состоянии.

2. КАК будет протестировано
3. КТО будет тестировать
4. ПОЧЕМУ это надо тестировать

Ошбика - то, что совершает человек (опечатка, человеческий фактор и т.д.)
Дефект - что-то, что не работает так, как оно должно работать. Дефект порождается ошибкой.
Отказ - нарушение работоспособности системы. Не любая ошибка может породить дефект. Не любой дефект может породить отказ. Отказ может быть внутренним и внешним. 

Баг или дефект репорт - это документ, описывающий ситуацию или последовательность действий, описывающий ситуацию или послежовательность действий, приведшую к некорректной работе. Баг-репорты с критическим приоритетом - всё надо бросить и бежать. Высокий приоритет - она серьёзная, но не блокирует бизнес-процесс. Средний приоритет - проблема не требует безотлагательных действий. Низкий приоритет - никто ничего не делает. 

Чит-листы - набор стандартных проверок, которые можно использовать во всех случаях жизни.



## Лекция 2

Виды тестирования: 
1. Функциональное тестирование - это способность ПО в определённых условиях решать задачи, нужные пользователям, согласно функциональным требованиям. Основывается на техническом задании. 
2. Нефункциональное тестирование. Относится к нему надежность, производительность, удобство использования, масштабируемость, безопасность продукта, отказ и восстановление. 

Принципы тестирования безопасности: 
1. Конфиденциальность- это сокрытие определенных ресурсов или информации. 
2. Целостность.
    - Доверие. Ресурс меняется только определенной группой пользователей.
    - Повреждение и восстановление. Насколько важной является процедура восстановления данных.
3. Доступность. Ограничение доступа широкому кругу.

Виды атак с точки зрения тестирования безопасности: 
1. XSS (Cross-Site Scripting)
2. XSRF / CSRF (Request Forgery) 
3. Authorization Bypass 
4. SQL-инъекции

Способы обезопасивания:
1. Проверка на экранирование символов (экранирование угловых скобок)
2. Замена опасных символов на их эквивалент в другой кодировке
3. Проверьте формирование url страницы, например, при авторизации (url не должен формировать персональные данные)
4. Проверить время жизни кук (слишком долго они жить не должны)

Проблемы UI/UX тестирования: 
1. Много разных устройств
2. Много разных предпочтений
3. Много окружений

Тесты для тестирования на отказ и восстановление: 
1. Отказ электричества на компьютер-сервер
2. Отказ электричества на компьютере-клиенте
3. Незавершенные циклы обработки данных
4. Объявление или внесение в массивы данных невозможных или ошибочных элементов
5. Отказ носителей данных
6. Потеря сигнала
7. Входящие звонки/смс
8. Переворот экрана

### Методы проектирования тестов: черный ящик. 

Основы: 
1. Области эквивалентных значений
    - Тестирование зависит от контекста
    - Нельзя проверить вообще всё
    Подразделяется на: 
    - Области для числового поля
    - Области для текста
    - Области для даты

2. Граничные значения
    На граничных значениях накосячить легче всего. Для этого важно определять скобочки (круглые и квадратные) для значений.
    Примеры: 
        (0;100). Тестовые кейсы: 0, 1, 99, 100
        (0; 100]. Тестовые кейсы: 0, 1, 100, 101
        [0: 100). Тестовые кейсы: -1, 0, 99, 100
        [0; 100]. Тестовые кейсы: -1, 0, 100, 101
    Принцип БМВ:
        - Большой. (сразу пол войны и мира)
        - Маленький (один символ)
        - В самый раз (средний размер)
    Границы текста: 
        - По длине
        - По состонию (пустое не пустое)
        - Не забываем про ноль 

    Типы границ: 
        - Логические (ограничения, накладываемые логикой)
        - Технологические (ограничение, накладываемой используемойтехникой)
        - Произвольная (ограничение, накладываемое аналитиком или разработчиком)
