# Факторы ранжирования сайтов:



1. Контент. 

Перед тем как оптимизировать контент, нужно понять: будет ли больше или меньше информации, чем на сайтах нашихконкурентов; почему люди будут читать наш текст; предлагаем ли мы ценность, в сравнении с нашими конкурентами?

Факторы качественного контента:
- Грамотный язык. Системы авто проверки: GoogleDocs, Яндекс.Спеллер, ORFO
- Верстка без ошибок. Системы проерки: валидатор w3
- УНикальный контент. Тексты нужно писать самостоятельно. Системы проверки: text.ru, http://www.content-watch.ru/text/

Ключевые слова - это то, как думают люди. С помощью данного инструмента мы можем видеть, как пользователь может увидеть нас сайт в поисковой системе, попасть на него. 

Ключевые фразы - фразы и слова, с помощью которых пользователь можем найти наш сайт. 

Правила использования: 
- Каждое ключевое слово должно быть минимум 5 раз на странице. 
- Но при этом сохранять здарый смысл. 
- Использовать только релевантные слова.
- СТраница не должна быть искусственной, она должна быть для людей. 

Типы запросов: 
- Высокочастотные
- Среднечастотные
- Низкочастотные

Набор ключевых фраз называется семантическим ядром. 

Дополнительные слова к семантическому ядру - это слова, которые дополняют основные ключевые фразы. 

Смежные темы -это дополнительные слова и запросы, не связанные напрямуюс темой, но связанные косвенно. 

Горизонтальный поиск - сбор данных со всего интернета. 

Вертикальный поиск - ориентирован на конкретный сегмент пользователей.

Но поисковики смотрят и на поведение пользователя на самой странице после клика. Их интересует:

1. Время, которое пользователь проводит на странице.
2. Показатель отказов — закрытие страницы сразу после открытия.
3. Глубина просмотров — переходы пользователя на другие страницы сайта. Если они есть, это хороший сигнал.
4. Действия — просмотры видео и клики по каким-то интерактивным элементам.
5. Переходы на другие страницы внутри домена.
6. Действия пользователя после закрытия страницы. Если он вернулся в выдачу и пошел щелкать другие запросы, значит на первом сайте он не нашел ответа — для поисковика это негативный показатель.
