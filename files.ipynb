```python
translate_dict = {"gasoline":"Бензин", 
                    "turbo gasoline":"Бензиновый турбонаддув",
                    "diesel":"Дизель",
                    "hybrid":"Гибридное топливо",
                    "electric":"Электроэнергия",
                    "gasoline turbo":"Бензиновый турбонаддув",
                    "hybrid gasoline turbo":"Гибридное топливо"
                    }
```

```python
dictMain = {}
for i in dirMain:
    dictMain[i] = {}

    pathTemp = pathMain + "\\" + i
    dirList = os.listdir(pathTemp)
    
    for n in range(len(dirList)):
        fullPath = pathTemp + "\\" + dirList[n]
        imgDir = os.listdir(fullPath)
        imgList =[]

        dirName = dirList[n]

        dirList[n] = dirName.split(', ')
        year = dirList[n][0].split()[-1]
        dirList[n][0] = ' '.join(dirList[n][0].split()[:-1])
        dirList[n].insert(1, year)
        carId = str(n+1) + ' ' + dirList[n][0]

        for temp in imgDir:
            imgList.append(i + "\\" + carId + "\\" + temp)

        if dirList[n][2].strip() == "electric":
            dictMain[i][carId]  = {"year":dirList[n][1],
                                "capacity":"",
                                "oil":dirList[n][2].lower(),
                                "power":dirList[n][3].lower().split("hp")[0],
                                "mileage":dirList[n][4].lower().split("km")[0],
                                "WD":dirList[n][5].upper(),
                                "img":imgList
                                }            

        else:    
            dictMain[i][carId]  = {"year":dirList[n][1],
                                "capacity":dirList[n][2].lower().split("cc")[0],
                                "oil":dirList[n][3].lower(),
                                "power":dirList[n][4].lower().split("hp")[0],
                                "mileage":dirList[n][5].lower().split("km")[0],
                                "WD":dirList[n][6].upper(),
                                "img":imgList
                                }
        if (dictMain[i][carId]["power"].isdigit() != True) and (dictMain[i][carId]["oil"] != "electric"):
            power = dictMain[i][carId]["oil"]
            dictMain[i][carId]["oil"] = dictMain[i][carId]["power"]
            dictMain[i][carId]["power"] = power

        dictMain[i][carId]["oil"] = translate_dict[dictMain[i][carId]["oil"]]
        #os.rename(fullPath, pathTemp + "\\" + carId)   

    print(i, dictMain[i])
```

```python
data = json.dumps(dictMain, indent=4, ensure_ascii=False)
print(data)
```
