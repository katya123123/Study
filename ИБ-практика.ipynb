## Практика 3

Классификаци сетей.

Тип коммутации. 
- Коммутация пакетов - всё что связано с компьютерными сетями. Это способ передачи данных. При поломке одного звена переопределяется путь и связь не разрывается. Работает задача маршрутизации.
- Коммутация каналов. Работает следующим образом: выбирается путь, при поломке пути ломается и соединение. Используется в сотовой связи.

Работа коммутации через систему сетевого оборудования. 

Технологии передачи: 
- Широковещательные сети. Мы всем посылаем информацию. ТВ и радио
- Точка-точка. Мы через узлы посылаем конкретно. Интернет

Протяженность. 
- Персональная (1 м)
- Локальная (10 м - 1 км)
- Муниципальная (10 км - 100 км)
- Глобальная (100 км - 1000 км)
- Объединение сетей (>1000 км)

Интернет - это объединение множества локальных сетей во едино. 

Топология сети - это конфигурация графа. Граф - это способ визуализации. Вершины - это узлы сети. Ребра - связи между узлами. 

Типы топологий: 
- Полносвязная (минусы: много проводов, нагрузка; плюсы: легко добраться до каждого узла)
- Ячеистая топология (минус: меньше проводов; плюсы: меньшепроводов)
- Звезда (По середине устройства, от него провода)
- Кольцо (Работает последовательно, то есть только в одном направлении; Минусы: долго, отказоустойчивость низкая; Есть на почте россии ) 
- Дерево (Отличиеот звезды в том, что звезда это базавая топология. У звезды одно сетевое оборудование, а у дерева - много. Сети вгуэса)
- Общая шина (отправилисообщение - оно пришло всем)
- Смешаная топология (это то как в мире устроено. Описывает иерархию устройств)

Виды топологий: 
- Физические (вот то что сверху рассмотрено) - соединение устройств в сети
- Логические - правила распростронения сигналов в сети 

Вай-фай: физически звезда логически шина

Классический эзернет - это технология передачи данных. Все подключено через вайфай

Сначала был коаксиальный кабель, который работал как шина физически и логически.
Потом появился Концентратор, которыйработал как звезда, но очень тупо (то есть как шина)

Виды сетевого оборудования: 
- концентратор. Всем подряд отправлял пакет.
- коммутатор. Коммутатор работает: к нему приходит пакет, читает пакет, понимает кому передать пакет, и отдает именно нужному компьютеру. Если такого нет устройства, то он отправит всем. Коммутатор не имеет выхода в интернет.
- маршрутизатор. Работает также, но чуть сложнее, есть выход в интернет и есть беспроводное подключение. Если не нашел устройство, куда отправить, то он не отправляется на широковещательный адрес. 

Коммутируемый эзернет - все подключено через провода. 

WiFi - физическая топология звезда (но только без провода), логически - общая шина. 

Де юро - формальные, юридические. Принятые по формальным законам. То, что установлено,  но не выполняется.
Де факто - фактические. Стандарт который реально выполняется, но юридически не закреплен. 

Консорциум W3C - стандарты web.

Стандарты IEEE:
802.3 эзернет
802.11 - вайфай беспроводные локальные сети
802.15 - блютуз персональные сети
802.16- широкополосные беспроводные сети 

Сложности: 
- Надежность 
- Развитие сети 
- распределение ресурсов
- безопасность
- качество обслуживания 

Решение:

Уровневая модель состоит из 4 абстрактны уровней, работает по принципу стека.

Концевик используется для отображения контрольных сумм (для восстановления пакета в случае утери). Основная цель концевика - использовать два вида алгоритмов: исправление ошибок и определени ошибок. 

Базовые поняти комп сетей: 
Протокол - правила соглашения, по которому работают два устройства.

Сервис - верхний уровень который мы получили в итоге

С помощью уровней можно понять что произошло на определнном уровне. 

Архитектура в сети - это набор уровней и протоколов в сети, интерфейс не входит в архитектуру. 
Стек протоколов- иерархически организованный набор протоколов, достаточный для организованных сетей

Эталонные модели организации сетей:

2. канальный. Разбиение 01 по пакетам. В конце пакета прикрепляется последовательность, которой не можем быть в пакете. Она отсекается. 
1. физический. Декодирование электрического сигнала в 01

Модель OSI - 7 уровней. 

Две основные технологии передачи: эзернет и вайффай. Если бы не было сетевого уровня, эзернет бы общался только с эзернетом

Обозначения: 
1. физический - бит. Декодирование в нули и единицы и из нулей и единиц в сигнал.
2. канальный - кадр. Обнаружение и коррекция ошибок + разбиение на кадры. Ошибки обнаруживаются по концевику (по хэш функции контрольной суммы)
3. сетевой - пакет. Решает две задачи: кому и как. Задачи: создание составной сети, согласование различий в сетях, адресация, маршрутизация. 
4. транспортный - сегмент. Обеспечивает передачу данных между процессами на хостах. 
5. СЕансовый уровень. Устанавливает сеансы связи. Основная задача: управление диалогом (кто в каком порядке будет отсылать сообщения). 
6. УРовень представления. Единый формат символов и чисел, шифрование и дешифрование. 
7. Прикладной уровень. Преобразование самим приложением в читабельный формат данных 

По http - не зашифровано, по https - зашифровано (http + tls)

1. ФИзический

Стек технологий - это структура по типу "первый вошел последний вышел"

# Практика 6 Криптографические методы

Включает два элемента: 
- криптография - шифрование
- стеганография - сокрытие самого факта передачи данных 

Криптоконтейнеры - это логическое пространство, в котором всё шифруется, что туда попадает 

pgp

Циммерман создал программу, которая объединяла в себе шифры. 

Эта программа - pgp. 

Типы шифров (шифры замены): 
- Атбаш. Шифр замены букв алфовита на букву алфовита с номером n - i + 1
- Шифр Цезаря. Сдвиг на n по алфовиту каждой буквы слова. Изначально было на 3, потом усложнили сдвиг на n. 
- шифр вижинера. Есть слово, есть ключ. Каждая буква сдвигается на ключ бук в алфавите

Типы шифров (шифры перестановки):
- Матрица из текста, столбцы меняются на строки или наоборот. Ключ - порядок столбцов (строк) 

Русский алгоритм криптографии - Кузнечик. Его прикол в ограниченной длине ключа. 

Шифр цезаря ломается морфологическим анализом. Избегается слитным написанием 
Ломается частотным анализом. Пытаемся заменить самые часто встречающиеся буквы на самые встречающиеся буквы в языке.

Системы шифрования: 
- с открытым ключом (симметричная) когда ключ шифровки = ключу дешифровки
- с закрытым  ключом (ассиметричная) когда ключ шифровки != ключу дешифровки 

Задача распределения ключей - задача для передачи ключа между пользователями.
