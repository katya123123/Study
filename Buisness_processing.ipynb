# Тема 1. Основы моделирования бизнесс-процессов 

Где применяется: 
1. переход на новые стандарты
2. новые продукты на рынок
3. автоматизация

Причины возникновения BPM (внешние): 
1. возросла доступность товаров и услуг
2. возросли тербовани потребителей
3. уменьшилось время жизни товаров

Причины внутренние: 
1. Рост сложности управленческих задач
2. Непригодность дальнейшего увеличесния числа сотрудников
3. недостаточная отдача 

Определения: 
1. Под моделироанием понимают процесс построения модели как образа оригинала, отбрадающее важнейшие черты и свойства
2. если модель уже построена,то моделирование - это процесс иссследования ыукнционирования системы или, вернее, её модли

Моделирование - это метод улучшения качетсва и эффективности работы организации. 
Моделирование описывает логическую взаимосвязь в рамках организации.

В более сложных ситуациях моделирование включает внешние процессы.
Моделирование позволяет провести анализ органиазации.

Цели моделирования бп: 
1. описание процессов. МОделирование позволяет получить "внешний" взгдяд на процессы.
2. нормирование процессов. моделирвоание задаёт правила выполнения процессов
3. установление взаимосвязей в процессах. устанавливает четкую связь между процессами и требованиями.

Определения: 
Хаммер, Чапли: "Совокупность различных видов деятельности в рамках которой на входе используется один или более видов резсурсов и в результате сощдаётся продукт представдяющий ценность для потребителя"

Основное определение: Моделирование - это нацеленноссть на результат необходимый клиенту посредством выполнения последовательности шагов (задач, действий, видов деятельности).

ПРоцессный подход:
1. кайдзен
2. Lean
3. 6 сигм
4. TQM
5. CMK

Преимущества процессного подхода: 
1. нацеленность сотрудников на результат процессов
2. быстрое реагирование на изменения
3. эффективное сокращение затрат

Принципы процессоного управления: 
1. Лидерство руководства
2. Вовлечение персонала
3. системный подход
4. эволюционность преобразований (изменения постепенны)
5. наличие цикла управления (планирование-> выполнение-> контроль -> актуализация)

Необходимость описания бизнес-процессов: 
1. выявлятя проблемные места
2. информация понятна для всех участников процесса
3. разграничение сфер ответственности между сотрудниками
4. позволяет увидеть движение документов и товарных ценностей
5. позволяет определить наиьроее значиные ресурсы для организации
6. упрощение проведения аудита для выявления необходимых изменений

Эталоны создаются в результате обощения соответствующего мирового опыта и существуют в форме международных стандартов.

- типовые процессы банка 

8 и 13 уровневые модели. Особенности.
- четкое агрегирование работ по результату 
- один процесс - один результат - одно подразделение -один бюджет - один владелец процесса 
- нужно мало персонала

8-ми процессная модель. Процессы: 
- выработка согласованных условий

- разработка и модификация продукта 
- воспроизводство средств производства 
- производство продукции 
- финансирование деятельности и расчеты по обязательствам 
- продвижение и продажи 

13-процессная. Процессы: 
- основные (изучение рынка, разработка стратегии, разработкапродуктов и услуг, маркетинг и продажи, производство и поставка услуг, выставление платежных требований)

Референтаная модель - это модель эффективного бизнес-процесса, созданная для предприятия кокнретной отрасли, внедренная на практике.

Операция - элементарное недилимое действие, выполняемое на одном рабочем месте
Функция - совокупность операций, сгрупиррованых по определенному признаку
Бизнесс-процесс - связанная совокупность функций, в ходе выполнения которой отребляются ресурсы и создается продукт, представляющий ценность для потребителя
Подпроцесс - бизнес-процесс, являющийся структурным элементом некоторого объемлющего бизнес-процесса

Описание бизнес процесса:
- определить владельца БП
- определить границы ответственности и полномочия владельца процесса 
- определить клиентов и выходы 
- определить поставщиков и входы 
- определить необходимые ресурсы 
- описать технологию выполнения БП 
- разработать показатели, по которым оценивается удовлетворенность клиентов БП
- описать работу владельца по анализу и улучшению БП

Владелец БП - тот, кто несет ответственность за результаты и эффективность БП, распоряжается инфраструктурой, персоналом и т.д.
Вход БП - продукт, который в ходе выполнения процесса преобразуется в выход,входы поступают извне.
Выход БП - результат выполнения БП. 
Ресурс БП - материальный или информационный объект, постоянно используемый для выполнения процесса, но не являющийся выходом процесса
Управление БП - документация, содержащая показатели норм, в соответствии с которыми осуществляется процесс 
Цель БП - необходимый или желаемый результат процесса 

Свойства БП: 
- Результативность характеризует соответствие результатов процесса ожиданиям потребителей 
- Определенность отражает степень, с которой реальный процесс соответствует описанию
- Управляемость характеризует степень, в которой произволится управление выполннием процесса производства
- Эффективность отражает насколько оптимально используютсяресурсы 
- Повторяемость характеризует способность процесса создавать выходные потоки содинаковыми характеристиками при повторных его реализациях 
- Гибкость адаптируемость способность процесса приспасабливаться к изменениям внешних условий 
- Стоимость определяет совокупную стоимость выполнения функций 

Владелец процесса: 
- определяет состав функций 
- последовательность работ впроцессе 
- показатели и критерии эфективности 
- организует мониторинг процесса 
- отвечает за мероприятия по повышению эффективности процесса 

Ландшафтная модель нужна для определения БП
Границы процесса - событие, инициирующее и завершающее процесс
Событие - наступление определенной ситуации 
Инициирующее событие - событие, при наступлении которого начинается процесс 
- Завершающее событие - событие, которым завершается процесс



# Лекция 4

Анализ процессов - это работа с графическими схемами, анализ всей доступной информации по процессам, измерения и их показателей, сравнительный анализ. 

Виды анализов БП: 
1. качественный анализ 
- качественный анализ процесса на основе субъективных оценок (SWOT-анализ, анализ проблем процесса, ранжирование процессов на основе субъективных оценок)
- визуальный качественный анализ
2. количественный анализ 



SWOT-анализ - анализ проесса предполагает выявление его сильных и слабых сторон, возможностей улучшения и угроз ухудшения. 
Четыре пункта описаны должны быть: 
Сильные стороны    Слабые стороны
Возможности        Угрозы

Как провести SWoT-анализ? 
- провести анкетирование руководителей и специалистов организации 
- обработать результаты анкетирования
- построить SWOT-анализ (таблицу)

Выделение проблемных областей - простейшее средства качественного анализа процесса.  Основное название - определенить направления дальнейшего более глубокого анализа. Для выявления необходимосформировать укрупненнуб схему процесса, отобразив на ней основные группы и функции. После этого указать на схеме проблемные области и их анализ. 


Ранжирование процессов на основе субъективной оценки. Ранжирование процессов выполняется на подготовительной стадии преокта. Методика: на первом этапе необходимо составить перечень основных процессов организации. Потом таблица со следющими столбцами: очень важный процесс, важный процесс, второстепенный процесс. И следующими строками: высокая эффективность, средняя эффективность, низкая эффективность. 

Визуальный анализ графических схем процесса. Сначала анализ входов и выходов состоит из двух ччастей: анализ потребностей во входах и выходах, анализ неиспользуемых выходов. Затем анализ ресурсов (по персоналу, инфраструктуре, ПО и т.д.) ПОзволяет выявить отсутствие необходимых функций, наличие излишних функций, дублирование функций.

Анализ процессов по отношению к типовым процессам. Любой процесс модно анализировать с точки зрения удовлетворения некоторым требованиям. Типовой процесс должен удовлетворять следующим процессам: регламентация всех составлюяющих процесса, использование цикла постоянного улучшения процесса. 

Измерение и анализ показателей процесса - важнейшие средства, позволяющие находить пути улучшения. Показатели: показатели процесса, показатели продкта процесса, показатели удовлетворенности клиента проуесса. 

Показатели времени вполнения процесс: 
- среднее время выполнения процесса в целом
- средснее время простоев 
- среднее время выполнения отдельных функций процесса 

Технические показатели процесса
- количество функций процесса, выполнеяемых на рабочих местаъ,
- численность персонала процесса
- те показатели, которые характеризуют технолонию выполнения процесса, используемое оборудование, ПО ит.д.

Показатели стоимости процесса: 
- стоимость процесса в целом 
- показатели стоимости процесса  (затраты на оплату труда исполнителей, амортизация, затраты на энергоносители, затраты на связь, затраты на получение информации, затраты на повышение квадификации и т.д.) 
- показатели стоимости продуктов процесса (стоимость сырья и материалов, затраты на оплату труда, амортизация оборудования, прочие затраты)

Методика ABC-анализа основана на: 
- определение ресурсов, используемых в процессах 
- опрелеоенеие операций процессов 
- определение объектов нечсения затрат
 - определние и расчет показателей связи ресурсы-операции
 - перенесении стоимости ресурсов на стоимость операции процесса
 - перенесение стоимости операций на стоимость готовых изделий 

 Стоимостные показатели: 
 - фонд зароботной платы 
 - затраты на энергоносители 
 - затраты на ремонт и техническое обслуживание 
 - потери от брака 
 - прочие
