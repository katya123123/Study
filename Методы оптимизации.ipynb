Лекция 1

W - переходный процесс
Ему на вход подаётся что-то, что можно постоянно замерять (х).
На выход выходит что-то, что также можно постоянно заммерять (у).
Z - это внешнее воздействие, внешние условия.
Данные о выходных значениях можно передать в устройство управления
ПИД-регулятор - это универсальное устройство управления.
