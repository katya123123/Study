## Лекция 2

Уровни носителей инфорации - это то, где мы можем увидеть информацию. Материальные, не материальные, человек.
Уровни взаимодействия - логический, синтаксический и семантический.

Признаки вещественного носителя инфы: 
1. инфа статична
2. может быть утрачена вместе с носителем

Низкоуровневое форматирование (логический уровень) - структура инфорамции на носителе, подразуменванет распределение на физическом носителе. После низкоуровневого форматирования на более высоких уровнях нельзя получить доступ к файлу.

Семантический уровень связан со смыслом передаваемой информации. 

Свойства информации дополнительные: 
1. концентрация (большие данные более секретные, чем одиночные)
2. рассеянность (обезличенная информация имеет низкую ценность)
3. целесообразность 

Жизненый цикл информации. Этапы: 
1. запись на носитель
2. выбрка информации
3. обработка информации
4. результаты обработки
5. оценка нужности информации
6. если не нужна, то удаляем 

В информационных ресурсах хранится задокументированная информация. 
Информационные ресурсы могут быть открытыми и ограниченного доступа. 

Нельзя делать ограниченным доступом: 
1. законы, нормативные акты
2. правовой статус государственных учреждений 
3. законы о ЧС
4. документы о деятельности гос властии и об использовании бюджетныхх средств

Предметные сферы тайн: 
1. природные (объективные)
2. людей (субъективные)

Гостайна - это защищаемая государством информация в различных сферах, распростронение такой информации может нанести ущерб государству. 

ФЗ от 1993.

Другие виды тайн не государственные. 

В NDA нельзя включать:
1. устав
2. учредительный документ
3. лицензии, удостоверения и патенты
4. формы отчетности по деятености 
5. сведения о платежеспособности
6. о численности работающих и зп и условиях труда
7. сведения об уплате налогов 
8. сведения об участии должностых лиц в предприятии

Признак секретного докумена - наличие сведений,составляющих гос тайну. 
Признак конфиденциального документа - содержание информации, подлежащей защите. 

Правовой режим информационных ресурсов определяет:
1. право собсвтеннности
2. категорию информации
3. порядок правовой защиты
4. порядок доступа к информации

Информация может быть выкуплена государством, если это может составлять государственную тайну. 

Защищенность информации контролирует государство. (ФСТЭК, Роскомнадзор, ФСБ)

ФСТЭК - федеральная служба по техническому и экспортному контролю. Она контролирует защиту информации. 

Все средства защиты подлежат сертификации в РФ. 

Персональные данные - любая информация прямо или косвенно относящаяся к определяемому лицу. 

Все информационные системы защиты должны быть аттестовваны. 

Процедурой проверки могут заниматься и коммерческие организации, и государство. 

## Лекция 3

ГСЗИ - гос система защиты информации
Система государственной безопасности. 
Государственная система обеспечения безопасности обеспечивает: 
1. Контроль за информацией 
2. Защиту от внутренних врагов

Виды национальной безопасности: 
1. транспортная 
2. государственная
3. информационная

Национальная безопасность - состояние защищенности жизненно важных национальных интересов от внутренних и внешних угроз и обеспечение конституционных прав граждан страны. 

Информационная безопасность - защищенность от внутренних и внешних информационных угроз.

Комплексный показатель национальной безопасности зависит от развития информационных ресурсов, от степени развитости нормативной базы.

Стратегия национальной безопасности - указ президента. 

Есть доктрина по информационной безопасности (от 2016 года).

Ещё документы об информации:
1. ФЗ об ЭЦП
2. Гражданский кодекс
3. Уголовный кодекс

Направления по защите информации:
1. гос. тайна
2. конфиденциальной информации (с 2000)
3. защита авторского права 
4. защита права на доступ к информации

Система нац безопасности: 
1. органы исполнительной власти 
2. государственные организации и объединения
3. граждане
4. законодательство

В любой гос организации должно быть лицо, ответственное за защиту информации.

КТо обеспечивает безопасность: 
1. ВС РФ
2. ФСИН
3. РКН
4. МВД
5. ФСБ
6. ФСТЭК
7. МЧС
8. Пограничные войска
9. ФТС
10. Совет по безопасности

ГСЗИ:
1. Законодительная  (федеральное собрание: совет федерации и госдума)
2. Исполнительная (президент: указы правительства)
3. Судебная (верховный суд: прокуратура)


Законы и постановления регулируют нормы защиты информации. 
Суды осуществляют контроль.

## Лекция 4

Угрозы системе безопасности. 
Моделирование системы безопасности: 
1. Описание угроз 
2. УГрозы конкретно для данной системы 
3. Этап  доступа к угрозам 
4. определение модели нарушителя/злоумышленника
5. определение соотношения стоимости информации и защиты информации 
6. ограничение предположений о возможных действиях нарушителя 

Типы угроз: 
- по типу природы 
- случайные и преднамеренные 
- по источникам угроз 
- по положению источника угроз (внутренние и внешние) 
- пассивные и активные  
- месторасположение информации 

Основные типы угроз: 
- угрозы целостности (преднамеренное изменение информации)
- угроза конфиденциальности (если информация становится известа другому лицу), уточка 
- угроза доступности (блокируется доступ в следствии действий злоумышленника)

Методы реализации угроз: 
- определить тип и паматеры носителя информации
- хищение машинных носителей информации 
- использование специальных средств перехвата сигнала (ПЭМИН - побочные электромагнитные излучения и наводки)
- несанкционированный доступ пользователя 
- несанкционированное превышение  пользователем своих полномочий 
- уничтожение информации 
- внесение изменений в программные компоненты 
- дезинформация 

Нарушитель - тот кто действует безумысла
Злоумышленник действует намеренно, из корыстных побуждений. 

Нарушители могут быть внутренние и внешние. 

Классификация нарушителей по уровню знания системы: 
- знание внутреннего устройства системы
- обладание высокого уровня знаний и опыта работы с техническими средствами 
- обладание высоким уровнем знаний в области программирования, высокий уровень владения системой 
- знание работы средств защиты, их способов 

Классификация по уровню возможностей: 
- запуск задач и программ, реализующие предусмотренные функции 
- запуск собственных программ реализующие другие функции 
- возможность управления функционированием системы 
- все возможности лиц, которые занимаются проектированием системы безопасности 

Все классификации иерархические. 

РД - руководящие документы 

Классификация по времени действия: 
- во время работы системы 
- в период неактивности компонентов системы 
- в процессе функционирования 

Классификация по месту действия: 
- без доступа на КЗ
- с КЗ без доступа в здание 
- внутри помещений но без доступа к техническим средствам 
- с рабочих мест конечных пользователей
- с доступам в зону обеспечения средств безопасности 

Модель нарушителя сравнивается с моделью информационной системы. Путем сравнения проводится оценка уязвимости. Оценка уязвимости выявляет актуальные для информационной системы.
