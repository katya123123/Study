## Лекция 4

СУБД - это специальный пакет программ с помощью которого реализуется централизованное управление базой данных и обеспечивается доступ к данным.
В состав СУБД входят трансляторы или интерпретаторы с языков описания или манипулярования данными. 

Позволяет администрировать, создавать модели и типы данных, создавать диаграммы БД и т.д.

SQL - язык манипулирования данными.
Есть ещё языки описания данных. 

ЯМД - языки манипулирования данными:
    - редактирование
    - добавить
    - удалить
    - выбрать 

ЯОД - язык описания данных, предназначен для описания схем базы данных, описание типов данных подлежащих хранению или выборке, их структур и связей.

Реляционная алгебра - когда работаем в целом с таблицами 
Реляционное исчисление - когда работаем с отдельными элементами таблицы

SQL основан на реляционных исчислениях.

Принцип работы СУБД:
1. Пользователь формирет запрос к БД на ЯОД.
2. СУБД принимает запрос и интерпретирует его 
3. СУБД анализирует по очереди схемы и отображения от внешней до внутренней, используя ЯМД
4. СУБД выполняет необходимые операции над хранимым данными и формирует внешнюю запись в направлении снизу вверх, которая в общем случае выдается пользователю на ЯОД.

Классификации СУБД: 
1. По языкам общения 
    - открытые 
    - закрытые
2. По числу поддерживаемых уровней
    - одноуровневые
    - двухуровневые
    - трехуровневые
3. По выполняемым функциям 
    - информационные 
    - операционные
4. По мощности 
    - настольные (локальные)
    - корпоративные (на любом предприятии). 
5. ПО сфере возможногоприменения
    - универсальные 
    - проблемно-ориентированные (это в основном 1С)
6. По категориям пользователей 
    - для разработчиков (чем ближе к БД, тем больше для разработчиков)
    - для конечных пользователей

### Современные модели данных.

1. Иерархическая модель. Выглядит как граф. Используется в разграничении прав доступа, в файловой системе, в системе документооборота. Это дерево реализовано как реляционная структура. Сейчас является частью других моделей, таких как реляционная, постреляционная или диаграмма классов. Корневые типы и их потомки. ПРедставляет собой упорядоченную совокупность экземпляров данных типа дерево.  Иерархическая модель использует меньше памяти за счет того, что не хранит дубли. Не предназначены для аналитики. 
2. Сетевая модель. Не ограничивает связи между записями, допускает тип связи многие-ко-многим, может включать в себя несколько иерархических структур. При этом структура сложная, возможны ошибки при вводе данных, как следствие высокое потребрелие ресурсов и сложность обработки данных. При вводе информации или её изменении нужно отслеживать.
3. Реляционная модель. Самая популярная. Предложена в компании IBM, сотрудник Кодд. СТруктура один-к-одному. Основные понятия: 
    1. Отношения - таблиц, совокупность множества данных, совокупность строк и столбцов. Множество упорядоченных строк (кортежей), таких что dn принадлежит Dn, где dn -строка (запись), Dn - домен, а n - степень отношений. 
    2. Домен - поле (столбец)
    3. Кортеж - запись (строка) 
    4. Ключ - подмножество атрибутов, которое уникально идентифицирует строку (кортеж). Свойства: 
       - УНикальность 
        - НЕ нулл
    5. Первичный ключ отличается допустимыми действиями: его нельзя обновлять и не может принимать значение нулл
    6. Внешний ключ - это ключ не первичный для одного отношения, но первичным для другого 
    Допустимые связи: 1:М, 1:1, М:1
    Нормализованное отношение- когда значение каждого атрибута в каждой строк недилимо, т е оно состоит из одного значения, а не из множеств. 
    Любая таблица в реляционной базе данных должна быть как минимум в первой нормальной форме обязательно. 
    Плюсы: 
        - простота, понятность
        - возможности обработки данных 
        
    Минусы: 
    
        - очень сложно реализовать сетевую или иерархическую модели

### 5. Многомерная модель. (Многомерные нереляционные)
Появилась после статьи в 1993 году Эдварда Кодда о многомерной модели. 
Особенности: 
- Историчность (хранение истории, привязка данных ко времени)
- Прогнозируемость (возможность использования математических функций прогнозирования)
- Агрегируемость (когда данные рассматриваются на разных уровнях обобщения)
Самое главное отличие - модель статична. Данные не обновляются. Они добавляются, но не обновляются. 
Измерение - это множество однотипных данных, образующих грани куба. 
Ячейка - это поле, значения которого однозначно определяются фиксированным набором измерений. 
Типы данных в ячейках как правило числовые.
Зачастую реляционные выгружают в многомерные модели. 
Если все кубики одинакового размера, то это гиперкубическая схема. 
Если кубики разного размера, то это поликубическая схема. 
Системы информационных систем: 
- OLTP (системы обработки транзакций, online transaction processing)
- OLAP (online analytic processing, СППР (система поддержки принятия решений))
СУБД:
- ORACLE Express Server (самая быстрая субд)
- MARKLOGIC SERVER (много мерная нереляционная модель данных, встроенное машинное обучение)
- ARANGODB (многопоточность, система обработки транзакций, позволяет организовать связь между кубами, свой язык AQL)
- OrientDB (поддерживает графы,документы, объекты, многомерные, реляционные. Поддерживает язык SQL. Позволяет импортировать реляционные бд. Позволяет организовывать распределенные ИС очень эффективно)
Специфические операции с многомерной моделью: 
1. Срез. Представление части куба (его подмножества) в виде двумерной таблички с данными, полученных в результате фиксации одного измерения. 
2. Вращение (rotate). Изменять последовательность измерений. 
3. Агрегирование (drill up). Детализация (drill down).

### 6. Объектно-ориентированная модель.

- Наследование. Распространяет область видимоти свойства на всех потомков объектов. 
- Полиморфизм. Способность одного и того же программного кода работать с разными типами данных.

В ооп-модели на сегодняшний день при выполнении отдельных операций позволяет использовать язык sql.

- возможность организации сложных связей между объектами.

- OLTP. Нацелены на быстрый ответ и многое количество пользователей, но при этом объемы информации не большие. Например банки, доставка, интернет-магазины, мфц, онайн-кассы. Проблемы: огромное количество запросов и обращений, сложно организовать защиту от DDOS. Необходимо несколько копий данных, при чем нужно понимать сколько копий нужно. 

Транзакция - некоторя законченная с точки зрения пользователя операци над бд. Любое обращение через SQL - транзакция. 

Свойства транзакций (ACID - стандарт соответствия этим свойствам): 
1. Атамарность - означает, что транзакция должна выполняться как единая операция доступа к данным (или выполнилась полностью, или не выполнилась совсем).
2. Согласованность - гарантирует взаимную целостность данных после окончания транзакции. 
3. Свойство изолированности - означает, что транзакции будут выполняться отдельно друг от друга и если транзакция не завершила операцию над данными, то остальным транзакциям доступ к этим данным запрещен. Для этого существует система блокировок. 
4. Долговечность - если транзакция была успешно выполнена, то произведенные ею изменения не будут потеряны ни при каких обстоятельствах. 

Уровни изоляции: 
- read uncommited - разрешает доступ к результатам выполнения ещё не зафиксированных транзакций и никак не ограничивает выполнение транзакций, тем самым допуская появление аномалий. Минимальный уровень изоляции. Позволяет считывать даже фантомы. Допустимо грязное чтение, повторяемое чтение и чтение фантомов..На русском - Предохраняющая блокировка от записей (кто первый запросил, того и тапка).
- read committed - результаты других транзакцций становятся доступными после их фиксаций, т.е. запрещает аномалия грязного чтения. Следующий уровень, фантомов больше нет. Допустимо повторное чтение и чтение фантомов.  На русском - Блокировка от полной блокировки. 
- repeatable read - повторное выполнение операций поиска и выбрки данных  дает такие же результаты, как первое. Нет ожидания выполнения, мы получаем данные на начало транзакции. Допустимочтение фантомов. На русском - Блокировка от записей. 
- serializable - требует, чтобы выполнение транзакций было эквивалентно некоторому последовательному выполнению. Сериализуемость - это максимальный уровень изолированности. Онгарантирует, что не будет меняться ни данные, ни структура данных до окончания транзакции. Не допустимо ничего,что карапнет данные. На русском - Полная блокировка.

Фиксация транзакции - это действие, обеспечивающее запись в БД всех изменений, которые были произведены в процессе её выполнения. До того, как транзакция зафиксирована, возможна отмена всех сделанных изменений. 

Журнал транзакций - один из механизмов, позволяющий фиксирлвать транзакции или производить откат той или иной транзакции. 

Откат - возвращение в исходное состояние. 

КОманды для завершения транзакци: 
- commit work 
- rollback work 

КОманды для создания контрольных точек: 
- savepoint

КОманды для начала транзакции: 
- set transaction 
- begin transaction

Пять основных команд для работы с транзациями в SQL и MySql.

### Информационная модель хранилища данных 

Многомерные хранилища данных. 
Две основные модели: 
- если все одинаковые - гиперкубическое хранилище 
- если все кубы разные - поликубическое 

ПЛюс - большая скорость обработки.
В таких моделях невозможно хранить большие объемы данных.

В таких кубиках данные, которые загружаются, уже должны быть аккумулированы. 
Для организации системы хранилища используяют реляционную модель. 
- В виде звезда: по центру факт, от них отходят справочные таблицы.
- Снежинка. Отличается тем, что каждый справочник детализирован за счет дополнительных описаний. Есть дополнительные таблицы, отходящие от справочников. 

Такая организация позволяет хранить большие объемы данных. Но при этом скорость обработки станет ниже. 

КИоск данных - специализированное тематическое хранилище, обслуживающее одно из направлений деятельности организации. (комбинация многомерного и реляционного подходов) 
В данном подходе существует проблема передачи данных между базой и рабочим местом.

### Документальные БД

Документальные БД работают только с документами. 

Обычная БД == фактографическая БД.

БД, которые хранят полные тексты - полнотекстовые.

БД, котораые хранят только ссылки - ссылочные. 

Репозиторий == словарь данных.

Типы документов:
- Пертинентные документы  - подходящий по смыслу поискового запроса. Соответствует смысловому запросу пользователя
- Релевантные документы - подходящие документы, соответствующие поисковому запросу по виду, но не обязательно по смыслу. Соотвествует содержанию информационного запроса пользователя

К документальным базам привязывается множество служебных баз для сжатия, обработки, хранения и тд.

Особенность документальных БД - они хранят и обрабатывают данные не имеющие четкой структуры и представленные в разных форматах. 

Документальная БД - бд,  в которой каждая запись отражает конкретный документ и содержит определенную информацию о нем. 

При работе с подобной базой документ может находится внутри неё, тогда это полнотекстовая (библиографическая, реферативная) бд. Или хранение только указателей на документы (БД-указатели). 

Поиск работает как БД-указатель. 

Информационный поиск зависит от информационного запроса пользователя. Запрос всегда выражен на естественном языке.
